```python
import torch
import requests
from PIL import Image

from diffusers import StableDiffusionDepth2ImgPipeline

pipe = StableDiffusionDepth2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-depth",
    torch_dtype=torch.float16,
).to("cuda")
```

```python
url = "https://img.freepik.com/free-photo/hopeful-man-asking-help-begging-favour-need-something-smiling-standing-yellow-background_1258-67086.jpg"
init_image = Image.open(requests.get(url, stream=True).raw)
prompt = "Japanese anime. Cute cartoon sticker. An young man is asking for something. Best quality. Extremely detailed. Cartoon styled. Not realistic."
n_prompt = "bad, deformed, ugly, bad anatomy"
image = pipe(prompt=prompt, image=init_image, negative_prompt=n_prompt, strength=0.7).images[0]
image
```
